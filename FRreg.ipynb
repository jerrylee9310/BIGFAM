```python
import pandas as pd
from BIGFAM import frreg
```

# Step 1. Load family data

```python
df_pair = pd.read_csv("test.info", sep='\t')
df_pheno = pd.read_csv("test.phen", sep='\t')
```

```python
df_pair.head()
```

```python
df_pheno.head()
```

```python
# merge info & phenotype
df_mrg = frreg.merge_pheno_info(df_pheno, df_pair)
df_mrg.head()
```

# Step 2. Degree-Of-Relatedness level

```python
df_frreg_DOR, _ = frreg.familial_relationship_regression_DOR(
    df_mrg,
    thred_pair=2, # hard-cut the small number of DOR pairs
    std_pheno=True,
    remove_outlier=True
)
```

```python
df_frreg_DOR
```

# Step 3. Relationship level

```python
df_frreg_REL, _ = frreg.familial_relationship_regression_REL(
    df_mrg,
    thred_pair=2, # hard-cut the small number of DOR pairs
    std_pheno=True,
    remove_outlier=True
)
```
