```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import sys
import sys; sys.path.append("/data/jerrylee/pjt/BIGFAM.v.0.1")

from BIGFAM import obj2, tools
import importlib
```

```python
source = "GS"
```

# Step 1. Load data

```python
# load FR-reg
frreg_path = f"/data/jerrylee/pjt/BIGFAM.v.0.1/data/{source}/frreg/REL"
fns = os.listdir(frreg_path)
len(fns)
```

# Step 2. Estimate X

```python
output_path = f"/data/jerrylee/pjt/BIGFAM.v.0.1/data/{source}/obj2"
```

## Step 2.1. X in all

```python
for ii, fn in enumerate(tqdm(fns)):
    pheno = fn.split(".")[0]
    print(f"[{ii+1}/{len(fns)}] {pheno}")
    
    # load FR-reg results
    df_frreg = pd.read_csv(f"{frreg_path}/{fn}", 
                           delim_whitespace=True)
    try:
        df_x = obj2.estimateX(df_frreg,
                            n_resample=100,
                            alpha_dicts={"type": "eta", "weight":-3},
                            regout_bin=["DOR", "sex_type"])
        
        if df_x.shape[0] > 0:
            # save raw results
            df_x.to_csv(f"{output_path}/{pheno}.X_raw",
                        sep='\t',
                        index=False)
            tools.raw2long(df_x, params=["eta", "alpha", "X"]).to_csv(
                f"{output_path}/{pheno}.X",
                sep='\t',
                index=False)
    except:
        print(fn)
```

## Step 2.2 X in male and female

```python
for ii, fn in enumerate(tqdm(fns)):
    pheno = fn.split(".")[0]
    print(f"[{ii+1}/{len(fns)}] {pheno}")
    
    # load FR-reg results
    df_frreg = pd.read_csv(f"{frreg_path}/{fn}", 
                           delim_whitespace=True)
    try:
        df_x = obj2.estimateXmXf(df_frreg,
                                alpha_dicts={"type": "eta", "weight":-2})
        if df_x.shape[0] > 0:
            # save raw results
            df_x.to_csv(f"{output_path}/{pheno}.XmXf_raw",
                        sep='\t',
                        index=False)
            df_x_sums = tools.raw2long(df_x, params=["eta", "alpha", "Xmale", "Xfemale"])
            
            df_x_sums.to_csv(
                f"{output_path}/{pheno}.XmXf",
                sep='\t',
                index=False)
            
            print(pheno)
            print(df_x_sums)
    except:
        print(fn)
        
```
