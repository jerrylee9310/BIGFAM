```python
df_res = pd.DataFrame(columns=[
    "cohort", "pheno", "BIGFAM", "lower_BIGFAM", "upper_BIGFAM"])

for fn in fns:
    cohort = fn.split(".")[0]
    pheno = fn.split(".")[1]
    reg_type = fn.split(".")[2]
    step = fn.split(".")[-1]
    
    if (reg_type == "intercept") & (step == "pred"):
        tmp = pd.read_csv(os.path.join(res_path, fn), sep='\t')
        
        # V(g)의 통계량 추출
        vg_median = tmp.loc[tmp["param"] == "V(g)", "median"].values[0]
        vg_lower = tmp.loc[tmp["param"] == "V(g)", "lower(2.5%)"].values[0]
        vg_upper = tmp.loc[tmp["param"] == "V(g)", "upper(97.5%)"].values[0]
        
        # df_res에 새로운 행 추가
        df_res.loc[len(df_res)] = [
            cohort, pheno, vg_median, vg_lower, vg_upper
        ]
```

```python
df_res.sort_values(by=["cohort", "pheno"]).to_csv(
    "/data/jerrylee/pjt/BIGFAM.v.2.0/data/other-methods/BIGFAM/BIGFAM.Vg.tsv",
    sep='\t',
    index=False
)
```

```python
df_res = pd.DataFrame(columns=[
    "cohort", "pheno", "BIGFAM", "lower_BIGFAM", "upper_BIGFAM"])

for fn in fns:
    cohort = fn.split(".")[0]
    pheno = fn.split(".")[1]
    reg_type = fn.split(".")[2]
    step = fn.split(".")[-1]
    
    if (reg_type == "intercept") & (step == "pred"):
        tmp = pd.read_csv(os.path.join(res_path, fn), sep='\t')
        
        # V(g)의 통계량 추출
        vg_median = tmp.loc[tmp["param"] == "V(g)", "median"].values[0]
        vg_lower = tmp.loc[tmp["param"] == "V(g)", "lower(2.5%)"].values[0]
        vg_upper = tmp.loc[tmp["param"] == "V(g)", "upper(97.5%)"].values[0]
        
        vs_lower = tmp.loc[tmp["param"] == "V(s)", "lower(2.5%)"].values[0]
        
        if vs_lower < 1e-4:
            continue
        
        # df_res에 새로운 행 추가
        df_res.loc[len(df_res)] = [
            cohort, pheno, vg_median, vg_lower, vg_upper
        ]
```

```python
p_to_exclude = [
    "Mean_platelet__thrombocyte__volume",
    "Platelet_count",
    "Heel_bone_mineral_density__BMD__T-score,_automated",
    "Standing_height", 
    "Impedance_of_leg__left_",
    "Impedance_of_leg__right_",
    "Immature_reticulocyte_fraction",
    "Neutrophill_percentage",
    "Lymphocyte_percentage",
    ]
df_res = df_res[~df_res["pheno"].isin(p_to_exclude)]
```
